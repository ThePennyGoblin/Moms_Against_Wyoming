```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
```

```python
browser = Browser("chrome")
```

```python
url = "https://www.fangraphs.com/roster-resource/injury-report?groupby=all&timeframe=all&season=2022"
```

```python
browser.visit(url)
```

```python
html = browser.html
```

```python
html_soup = soup(html, "html.parser")
```

```python
browser.links.find_by_partial_text("Single Grid").click()
```

```python
dropdown = browser.find_by_tag('select')
```

```python
dropdown.select_by_text('Infinity')
```

```python
# Initialize the DataFrame list
injury_table = []
# Loop through all pages
while True:
    # Parse the HTML
    html = browser.html
    html_soup = soup(html, "html.parser")
    
    # Find the table and rows
    tables = html_soup.find_all('table')
    rows = tables[15].find_all('tr')  # Adjust the index if necessary
    
    # Extract headers and data
    if len(injury_table) == 0:
        headers = [x.text for x in rows[0].find_all('th')]
    
    for row in rows[1:]:
        injury_table.append([val.text for val in row.find_all("td")])
    
    # Check if there's a "Next" button and click it, otherwise break
    next_button = browser.links.find_by_partial_text("Next")
    if next_button:
        next_button.click()
    else:
        break

# Convert the list to a DataFrame
df = pd.DataFrame(injury_table, columns=headers)
df
```
